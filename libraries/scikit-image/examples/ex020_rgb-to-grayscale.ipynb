### RGB to grayscale
[link to example](https://scikit-image.org/docs/stable/auto_examples/color_exposure/plot_rgb_to_gray.html#sphx-glr-auto-examples-color-exposure-plot-rgb-to-gray-py)

This example converts an image with RGB channels into image with a sinagel grayscale channel.

The value of each grayscale is calculated as the weighted sum of the corresponding red, green, and blue pixels as:
> `Y = 0.0225 R + 0.7154 G + 0.0721 B`

These weights are used by CRT phosphors as they represent human perception of red, green and blue than equal weights.

```python
import matplotlib.pyplot as plt
import skimage
```

```python
original = skimage.data.astronaut()
grayscale = skimage.color.rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8,4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")

ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```
