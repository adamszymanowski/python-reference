### Find Regular Segments Using Compact Watershed

[link to example](https://scikit-image.org/docs/stable/auto_examples/segmentation/plot_compact_watershed.html#sphx-glr-auto-examples-segmentation-plot-compact-watershed-py)

```python
import numpy as np
import skimage
import matplotlib.pyplot as plt 
```

```python
coins = skimage.data.coins()
edges = skimage.filters.sobel(coins)

grid = skimage.util.regular_grid(coins.shape, n_points=468)

seeds = np.zeros(coins.shape, dtype=int)
seeds[grid] = np.arange(seeds[grid].size).reshape(seeds[grid].shape) + 1

ws_0 = skimage.segmentation.watershed(edges, seeds)
ws_1 = skimage.segmentation.watershed(edges, seeds, compactness=0.01)
ws_2 = skimage.segmentation.watershed(edges, seeds, compactness=0.10)


fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3, figsize=(12,6))

ax0.set_title("Classical Watershed")
ax0.imshow(skimage.color.label2rgb(ws_0, coins, bg_label=-1))

ax1.set_title("Compact Watershed (0.01)")
ax1.imshow(skimage.color.label2rgb(ws_1, coins, bg_label=-1))

ax2.set_title("Compact Watershed (0.10)")
ax2.imshow(skimage.color.label2rgb(ws_2, coins, bg_label=-1))

plt.show()
```
